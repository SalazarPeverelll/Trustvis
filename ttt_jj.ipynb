```python
import torch
import os
import matplotlib.pyplot as plt
import numpy as np

tensor_values = np.load('probs.npy')

plt.hist(tensor_values, bins=100)
# plt.xlabel('Tensor Value')
plt.ylabel('Frequency')
# plt.title(f'Distribution of wij')
plt.show()
```

```python
from scipy.stats import rankdata
rank_transformed_weights = rankdata(tensor_values, method='average') / len(tensor_values)

rank_transformed_weights = rank_transformed_weights+ 0.04
plt.hist(rank_transformed_weights, bins=35)
# plt.xlabel('Tensor Value')
plt.ylabel('Frequency')
# plt.title(f'Distribution of wij')
plt.show()
```
